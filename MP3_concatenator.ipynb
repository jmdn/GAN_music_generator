```
clip_stack, sample_rate = torchaudio.load('/content/drive/My Drive/Colab Notebooks/Mp3_stuff/training_mp3s/generated_clip_batchnorm_0.mp3')
x = 0
while x < 8240:
  clip, sample_rate = torchaudio.load('/content/drive/My Drive/Colab Notebooks/Mp3_stuff/training_mp3s/generated_clip_batchnorm_' + str(x) + '.mp3')
  clip_stack = torch.cat((clip_stack,clip),1)
  print(clip_stack.size())
  x = x + 10

torchaudio.save('/content/drive/My Drive/Colab Notebooks/Mp3_stuff/training_mp3s/stacked_generated_clip_batchnorm_.mp3', clip_stack, sample_rate)
        
        
        
```
